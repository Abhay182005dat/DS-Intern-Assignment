####Lets visualize to gain some knowledge or to get some ideas on to choose or drop unwanted features for us

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['equipment_energy_consumption'], kde=True, bins=10)
plt.title('Distribution of Equipment Energy Consumption')
plt.xlabel('Energy Consumption (Wh)')
plt.ylabel('Frequency')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(9))
plt.grid(True)
plt.show()
```
